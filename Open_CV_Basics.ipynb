```python
#Split the images into different color index using split image cv2.split()
import numpy as np
import cv2
image =cv2.imread(r"C:\Users\Dipak\Desktop\Rea\IMG_6958.JPG")
print(image.shape)
B,G, R = cv2.split(image)
print(B)
cv2.imshow("Red",R) # All images would be gray as we have one dimension
cv2.imshow("Green",G)
cv2.imshow("Blue",B)
cv2.waitKey(0)


#Merge the image will get it to original image
Merged = cv2.merge([R,G,B])
cv2.imshow("Merged",Merged)
cv2.waitKey(0)

#Want to amplify blue color
Merged= cv2.merge([R+100,G,B])
cv2.imshow("merge the image and amplify blue by 100 max it can go to 255",Merged)
cv2.waitKey(0)


# Want to make image of Red, Green, & Blue 
#Let's create a matrix of zeros
#with dimensions of the image h x w
#here actually the chanel (3024, 4032, 3) that is 3 R,G,B so its one dimension we are converting to 3 channel but replacing the colors with zeros to convert so only one color is used and others are zeros
zeros = np.zeros(image.shape[:2], dtype ="uint8")
cv2.imshow("RED ONLY", cv2.merge([zeros,zeros,R]))
cv2.imshow("GREEN ONLY", cv2.merge([zeros,G,zeros]))
cv2.imshow("BLUE ONLY", cv2.merge([B,zeros,zeros]))
cv2.waitKey(0)
```
